## <center><font color=tomato>Multi-class Classification</font></center>
### <font color=lightGreen>Dataset</font>

`ex3data1.csv` that contains 5000 training examples of handwritten digits

- Each training example is a 20 pixel by 20 pixel grayscale image of the digit.
- Each pixel is represented by a floating point number indicating the grayscale intensity at that location.
- The 20 by 20 grid of pixels is 'unrolled' into a 400-dimensional vector.
- Each of these training examples becomes a single row in our data matrix X.
- This gives us a 5000 by 400 matrix X where every row is a training example for a handwritten digit image.


```python
from numpy import loadtxt
from display_data import displayData
import matplotlib.pyplot as plt
from logistic_regression import *
import numpy as np


data = loadtxt("data/ex3data1.csv", delimiter=",", skiprows=1)
X = data[:, :-1]
y = data[:, -1]
```

The second part of the training set is a 5000-dimensional vector y that contains labels for the training set 0 to 9.

## <center><font color=tomato>Visualizing the data</font></center>
The code below randomly selects selects 100 rows from X and passes those rows to the displayData function.

This function maps each row to a 20 pixel by 20 pixel grayscale image and displays the images together.

```python
m = X.shape[0]
rand_indices = np.random.permutation(m)
sel = X[rand_indices[0:100], :]
displayData(sel);
```

## <center><font color=tomato>All-vs-one classification</font></center>
### <font color=lightGreen>All-vs-one prediction</font>

```python
Lambda = 0.2
```

```python
theta_opt = multi_class_theta_optimization(X, y, Lambda, gradient_decent=True, iterations=300, alpha=4)
acc = multi_class_prediction_accuracy(X, y, theta_opt)
print(f'Training Set Accuracy with gradient decent: {acc*100:.2f}%')
```
